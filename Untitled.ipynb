```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def pltImage(img):
    plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
    plt.xticks([]), plt.yticks([]) # to hide tick values on X and Y axis
    plt.show()
```

```python
#img > -  > + > + > -
```

```python
img = cv2.imread('finger.jpg',0)
out = img.copy()
out = 255 - out
pltImage(out)
```

```python
kernel = np.ones((3,3),np.uint8)
erosion = cv2.erode(out,kernel,iterations = 1)
dilation = cv2.dilate(erosion,kernel,iterations = 1)
dilation2 = cv2.dilate(dilation,kernel,iterations = 1)
erosion2 = cv2.erode(dilation2,kernel,iterations = 1)


#opening = cv2.morphologyEx(out, cv2.MORPH_OPEN, kernel)
#closing = cv2.morphologyEx(opening, cv2.MORPH_CLOSE, kernel)
```
